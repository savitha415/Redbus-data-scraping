```python
import pandas as pd

df1 = pd.read_csv("bus_data.csv")

df1
```

```python
df1.drop(columns=['Link'])
```

```python
df1.info()
```

```python
df1['Price'] = df1['Price'].astype(str)
df1['Price'] = df1['Price'].str.replace('INR ', '').astype(float)
df1['Price'].fillna(0)
```

```python
df1['Star Rating'] = df1['Star Rating'].str.split('\n').str[0]
df1['Star Rating'].fillna(0)
```

```python
df1.isnull().sum()
```

```python
# Function to safely convert to float
def safe_float(value):
    try:
        return float(value)
    except ValueError:
        return None  # or another appropriate value, such as NaN

# Clean 'Star Rating' column
df1['Star Rating'] = df1['Star Rating'].apply(safe_float)
```

```python
df1 = df1.dropna(subset=['Bus Name'])
df1 = df1.dropna(subset=['Bus Type'])
df1 = df1.dropna(subset=['Departing Time'])
df1 = df1.dropna(subset=['Reaching Time'])
df1 = df1.dropna(subset=['Duration'])
df1 = df1.dropna(subset=['Star Rating'])
df1 = df1.dropna(subset=['Price'])
df1 = df1.dropna(subset=['Seats Availability'])
```

```python
df1.isnull().sum()
```

```python
import re
def extract_number(seat_str):
    match = re.search(r'\d+', seat_str)  # Find the first sequence of digits
    if match:
        return float(match.group(0))
    return 0  # Default value if no number is found
```

```python
df1['Seats Availability'] = df1['Seats Availability'].apply(extract_number)
```
